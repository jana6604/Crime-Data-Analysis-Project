```python
df['Date occurred'] = pd.to_datetime(df['Date occurred'])
Daily_crime = df.groupby(df['Date occurred'].dt.date).size()

plt.figure(figsize=(10,8))

Daily_crime.plot(kind='line', color='darkred', linewidth=1.5)
plt.title('"Daily Crime Trends[2020 - 2024]  (Line plot)"', fontsize=16, fontweight='bold', color='navy')
plt.xlabel('"Date (YYYY-MM)"', fontsize=12, fontweight='bold', color='navy')
plt.ylabel('"Number of Crimes"', fontsize=12, fontweight='bold', color='navy')      
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
Daily_crime.index = Daily_crime.index.to_timestamp()
```

```python
df['Date occurred'] = pd.to_datetime(df['Date occurred'])
Daily_crime = df.groupby(df['Date occurred'].dt.to_period('M')).size()

plt.figure(figsize=(12,5))

Daily_crime.plot(kind='line', color='Green', marker='o')
plt.title('"Monthly Crime Trends [2020-2024] (line plot)"', fontsize=16, fontweight='bold', color='navy')
plt.xlabel('"Months by Years"', fontsize=12, fontweight='bold', color='navy')
plt.ylabel('"Number of Crimes"', fontsize=12, fontweight='bold', color='navy')      
plt.grid(True, linestyle='--', alpha=0.8)

plt.show()
```
